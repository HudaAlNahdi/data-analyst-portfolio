### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ تنظيف البيانات
df_clean = df['ConvertedCompYearly'].copy()
df_clean = pd.to_numeric(df_clean, errors='coerce')
df_clean = df_clean.dropna()

# 2️⃣ رسم Histogram
plt.figure(figsize=(10,6))
plt.hist(df_clean, bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Yearly Compensation (ConvertedCompYearly)')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('Number of Respondents')
plt.yscale('log')  # لو حبيت التعامل مع القيم الكبيرة جدًا يمكن تفعيل اللوغاريتمية
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ تحويل الفئات العمرية إلى أرقام
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70,
    'Prefer not to say': None
}

df['AgeNumeric'] = df['Age'].map(age_map)

# 2️⃣ تنظيف البيانات
df_clean = df.dropna(subset=['AgeNumeric']).copy()

# 3️⃣ رسم Box Plot
plt.figure(figsize=(8,5))
plt.boxplot(df_clean['AgeNumeric'], vert=True, patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='blue'),
            medianprops=dict(color='red'))
plt.title('Box Plot of Age')
plt.ylabel('Age (Numeric)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
import pandas as pd
import matplotlib.pyplot as plt

# التأكد من وجود العمود الرقمي للعمر
if 'AgeNumeric' not in df.columns:
    age_map = {
        'Under 18 years old': 16,
        '18-24 years old': 21,
        '25-34 years old': 30,
        '35-44 years old': 40,
        '45-54 years old': 50,
        '55-64 years old': 60,
        '65 years or older': 70,
        'Prefer not to say': None
    }
    df['AgeNumeric'] = df['Age'].map(age_map)

# تنظيف البيانات
df_clean = df.dropna(subset=['AgeNumeric', 'ConvertedCompYearly']).copy()
df_clean['ConvertedCompYearly'] = pd.to_numeric(df_clean['ConvertedCompYearly'], errors='coerce')
df_clean = df_clean.dropna(subset=['ConvertedCompYearly'])

# رسم Scatter Plot مع محور Y لوغاريتمي
plt.figure(figsize=(10,6))
plt.scatter(df_clean['AgeNumeric'], df_clean['ConvertedCompYearly'], alpha=0.5, color='green')
plt.title('Scatter Plot of Age vs ConvertedCompYearly (Log Scale)')
plt.xlabel('Age (Numeric)')
plt.ylabel('ConvertedCompYearly')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
import pandas as pd
import matplotlib.pyplot as plt

# التأكد من العمود الرقمي للعمر
if 'AgeNumeric' not in df.columns:
    age_map = {
        'Under 18 years old': 16,
        '18-24 years old': 21,
        '25-34 years old': 30,
        '35-44 years old': 40,
        '45-54 years old': 50,
        '55-64 years old': 60,
        '65 years or older': 70,
        'Prefer not to say': None
    }
    df['AgeNumeric'] = df['Age'].map(age_map)

# تنظيف البيانات
df_clean = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'AgeNumeric']).copy()
df_clean['ConvertedCompYearly'] = pd.to_numeric(df_clean['ConvertedCompYearly'], errors='coerce')
df_clean['JobSatPoints_6'] = pd.to_numeric(df_clean['JobSatPoints_6'], errors='coerce')
df_clean['AgeNumeric'] = pd.to_numeric(df_clean['AgeNumeric'], errors='coerce')
df_clean = df_clean.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'AgeNumeric'])

# أخذ عينة صغيرة (مثلاً 10% من البيانات)
df_sample = df_clean.sample(frac=0.1, random_state=42)

# رسم Bubble Plot
plt.figure(figsize=(12,8))
bubble_size = df_sample['AgeNumeric'] * 5
plt.scatter(df_sample['ConvertedCompYearly'], df_sample['JobSatPoints_6'],
            s=bubble_size, alpha=0.5, color='purple', edgecolors='w', linewidth=0.5)
plt.title('Bubble Plot: Compensation vs Job Satisfaction with Age as Bubble Size (Sample)')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('JobSatPoints_6')
plt.xscale('log')  # لأن التعويض واسع الانتشار
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ تنظيف البيانات (التأكد من وجود قيم صالحة)
df_clean = df['MainBranch'].dropna().copy()

# 2️⃣ حساب عدد كل فئة
counts = df_clean.value_counts().sort_values()  # ترتيب من الأقل للأكثر لتسهيل الرسم الأفقي

# 3️⃣ رسم Horizontal Bar Chart
plt.figure(figsize=(10,6))
counts.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title('Distribution of Respondents by Main Branch')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ تنظيف البيانات
df_clean = df['LanguageWantToWorkWith'].dropna().copy()

# 2️⃣ حساب تكرار كل لغة
language_counts = df_clean.value_counts()

# 3️⃣ اختيار أعلى 5 لغات
top5_languages = language_counts.head(5)

# 4️⃣ رسم Vertical Bar Chart
plt.figure(figsize=(8,6))
top5_languages.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Top 5 Programming Languages Respondents Want to Work With')
plt.xlabel('Programming Language')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ تنظيف البيانات
df_clean = df.dropna(subset=['Age', 'JobSatPoints_6', 'JobSatPoints_7']).copy()
df_clean['JobSatPoints_6'] = pd.to_numeric(df_clean['JobSatPoints_6'], errors='coerce')
df_clean['JobSatPoints_7'] = pd.to_numeric(df_clean['JobSatPoints_7'], errors='coerce')
df_clean = df_clean.dropna(subset=['JobSatPoints_6', 'JobSatPoints_7'])

# 2️⃣ ترتيب فئات العمر (اختياري، لجعل الرسم مرتب)
age_order = ['Under 18 years old', '18-24 years old', '25-34 years old',
             '35-44 years old', '45-54 years old', '55-64 years old',
             '65 years or older', 'Prefer not to say']

# 3️⃣ تجميع البيانات حسب العمر وحساب الوسيط لكل JobSatPoints
median_by_age = df_clean.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()
median_by_age = median_by_age.reindex(age_order)  # ترتيب الفئات حسب العمر

# 4️⃣ رسم Stacked Bar Chart
plt.figure(figsize=(10,6))
median_by_age.plot(kind='bar', stacked=True, color=['skyblue','orange'])
plt.title('Median Job Satisfaction by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Job Satisfaction Points')
plt.xticks(rotation=45)
plt.legend(title='JobSatPoints')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ تنظيف البيانات
df_clean = df['DatabaseHaveWorkedWith'].dropna().copy()

# 2️⃣ تقسيم القيم المدمجة (إذا كانت مفصولة بفواصل) وتحويلها إلى قائمة
all_dbs = df_clean.str.split(';').explode()  # استخدام ';' إذا كانت القيم مفصولة بفواصل من النوع ; أو ',' حسب بياناتك
all_dbs = all_dbs.str.strip()  # إزالة الفراغات

# 3️⃣ حساب التكرارات لكل قاعدة بيانات
db_counts = all_dbs.value_counts()

# 4️⃣ اختيار أعلى 10 قواعد بيانات (يمكن تعديل الرقم حسب الرغبة)
top10_dbs = db_counts.head(10)

# 5️⃣ رسم Bar Chart
plt.figure(figsize=(10,6))
top10_dbs.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Top 10 Databases Respondents Have Worked With')
plt.xlabel('Database')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ تنظيف البيانات
df_clean = df.dropna(subset=['Age', 'ConvertedCompYearly']).copy()
df_clean['ConvertedCompYearly'] = pd.to_numeric(df_clean['ConvertedCompYearly'], errors='coerce')
df_clean = df_clean.dropna(subset=['ConvertedCompYearly'])

# 2️⃣ ترتيب فئات العمر
age_order = ['Under 18 years old', '18-24 years old', '25-34 years old',
             '35-44 years old', '45-54 years old', '55-64 years old',
             '65 years or older', 'Prefer not to say']

# 3️⃣ تجميع البيانات وحساب الوسيط لكل فئة عمرية
median_by_age = df_clean.groupby('Age')['ConvertedCompYearly'].median().reindex(age_order)

# 4️⃣ رسم Grouped Bar Chart
plt.figure(figsize=(10,6))
median_by_age.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Median ConvertedCompYearly by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median ConvertedCompYearly')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
